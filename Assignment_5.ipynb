#### IV. Detect and treat outliers if necessary.

```python
data.boxplot(figsize=(12,5))
plt.show()
```

#### According to above Boxplot, yes outliers are present in dataset. Soo we have to remove or replace this outliers

```python
for col in data.columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    outliers = data[(data[col]< lower) | (data[col]>upper)]
    #print("No of outliers in ",col,":",outliers.shape[0])
    print(f"{col} : {outliers.shape[0]} outliers")
```

#### According to above report there are much count of outliers soo we can't remove this data.. 
#### If we remove it we will lost much data, soo we will replace the data with upper and lower bound values

```python
for col in data.columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    data.loc[data[col] < lower, col] = lower
    data.loc[data[col] > upper, col] = upper
```

```python
data.boxplot(figsize=(12,5))
plt.show()
```

#### After replacing the outliers we can see the boxplot... yes now we have cleaned data

### 2. Statistical Summary:

#### Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).

```python
print("Stastical Summary") 
summary = []
for col in data.columns:
    mean = data[col].mean()
    median = data[col].median()
    std = data[col].std()
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    summary.append([col,mean,median,std,IQR])
summary_table = pd.DataFrame(summary,columns=["Columns",'Mean','Median','Std','IQR'])
summary_table                             
```

### Highlight any interesting findings from this summary.

#### 1. In LB Attribute Mean ≈ 133.29 and Median = 133 Since mean and median are almost equal, the data is fairly symmetric with little skewness.

#### 2. we can see in AC, FM, UC, DL variables have very small mean and median values close to zero. This suggests that these eventsoccur rarely in the dataset.

#### 3. DS, DP variables shows all the values are zero.. soo this attributes will noot affect to the table.. it is not usefull data

#### 4. In ASTV Mean ≈ 46.98 and Standard Deviation ≈ 17.61.
#### A relatively high standard deviation and IQR (29) indicates large variability in this feature.

#### 5. In ALTV (Abnormal Long-Term Variability) Median = 0, but Mean ≈ 6.69.
#### This suggests the data may be right-skewed, meaning a few high values are increasing the mean.

#### 6. Width Variable Mean ≈ 70.32 and Standard Deviation ≈ 39.57.

#### 7. NSP (Target Variable) Mean = 1 and Standard Deviation = 0.
