#### 6. Given a 'marks' variable, print the grade:
#### - 90 and above: 'A'
#### - 75-89: 'B'
#### - 50-74: 'C'
#### - Below 50: 'Fail'

```python
marks = int(input("Enter marks:"))
if marks > 90:
    print("Student Pass with Grade : A")
elif marks > 75 & marks < 89:
    print("Student Pass with Grade : B")
elif marks > 50 & marks <74:
    print("Student Pass with Grade : C")
elif marks < 50:
    print("Student is Fail")
else:
    print("Please enter valid percentage")
```
