#### Calculate the mean, median, mode, and standard deviation for these columns.
#### Provide a brief interpretation of these statistics.

```python
Mean_Value = df[num_col].mean() # Mean function use to find average value
print("Mean Values :\n",Mean_Value)
```

```python
Median_value = df[num_col].median() # Find Middle Values when data is sorted
print("Median Values :\n",Median_value)
```

```python
Mode_value = df[num_col].mode() # The most frequently occurring value.
print("Mode Values :\n",Mode_value)
```

```python
Stnd_Dev = df[num_col].std() # Low standard deviation means data is close to mean high means data is spread
print("Mode Values :\n",Stnd_Dev)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.hist(df['Volume'],color='purple',edgecolor='k',alpha=0.8)
plt.xlabel('Volume')
plt.ylabel("Frequency")
plt.title("Distribution of Volume")
plt.show()
```

#### Box Plots:
##### Create boxplots for numerical variables to identify outliers and the interquartile range.
##### Discuss any findings, such as extreme values or unusual distributions.

```python
for col in num_col:
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,2)
    plt.boxplot(df[col],vert=False)
    plt.title(f"Boxplot of {col}")
plt.show()
```

```python
Q1 = df['Volume'].quantile(0.25)
Q3 = df['Volume'].quantile(0.75)

IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Volume'] < lower_bound) | (df['Volume'] > upper_bound)]
outliers['Volume']
```

 ### Bar Chart Analysis for Categorical Column:
 #### Identify categorical columns in the dataset.
 #### Create bar charts to visualize the frequency or count of each category.
 #### Analyze the distribution of categories and provide insights.

```python
cat_col = df.select_dtypes(include=['object']).columns
cat_col
```

```python
for col in cat_col:
    df[col].value_counts().plot(kind='bar',color = '#800060')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    #ax = plt.bar(cat_col[col],cat_col['count'],color = '#98FF98')
    #plt.bar_label(ax)
    plt.show()
```

### Analyze the distribution of categories and provide insights.

### Conclusion
#### According to above bar chart in Date, SKU,City, BU and Model columns shows that all categories have nearly equal counts.
#### There is no single category that dominates the dataset. 
#### This indicates that the data is evenly distributed across the categories, similar to a uniform distribution.

#### For Day And model columns the one category has higher count. The bars are not equal but mostly close, except one higher bar.
#### This indicates a Skewed Distribution, because one category has a much higher frequency than the others, making the data uneven.
