### Exercise 1:
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python

n= int(input("Enter Number:"))
if n==1 | n<0:
    print("Number is not prime number")
if n>1:
    for i in range(2,n):
        if n%2==0:
            print("Number is not prime")
            break
    else:
        print("Number is prime")
    
```

### Exercise 3: 
#### Squares of Even/Odd Numbers 
Create a python script that prints the squares of all even or odd numbers within the
range of 100 to 200. Choose eighther even or odd nembers and documents your choice in the code

```python
#create two list to store even and odd values And two list to store there Squares seperatly
even=[]
SqEven=[]
odd=[]
SqOdd=[]
ch=int(input("Enter your choice:\n 1. Even\n 2. Odd\n")) # Ask to the user which numbers u want to see.
for i in range(100,201): # to check the number from 100 to 200, I take here range to 201 to check the last number also
    if i%2==0:
        even.append(i) # using append fuction add the values in list
        SqEven.append(i*i)
    else:
        odd.append(i)
        SqOdd.append(i*i)
if ch==1:          # according to the users choice it will shows Even Or Odd numbers List
    print("Even Numbers And their Squares from 100 to 200:\n")
    print("Numbers       Squares")
    print("---------------------")
    for i in range(len(even)):
        print(even[i], "         ",SqEven[i])
elif ch==2:
    print("Odd Numbers And their Squares from 100 to 200:\n")
    print("Numbers       Squares")
    print("---------------------")
    for i in range(len(odd)):
        print(odd[i], "         ",SqOdd[i])
else:
    print("Enter valid choice..") # if the user enter number else from 1 and 2 then it will shows this msg
```
